```python

import random


# Function to add Gaussian noise to the image
def add_gaussian_noise(image, mean=0, stddev=25):
    gaussian_noise = np.random.normal(mean, stddev, image.shape)
    noisy_image = image + gaussian_noise
    noisy_image = np.clip(noisy_image, 0, 255)  # Ensuring pixel values remain within [0, 255]
    return noisy_image.astype(np.uint8)

# Function to add salt-and-pepper noise to the image
def add_salt_pepper_noise(image, salt_prob=0.02, pepper_prob=0.02):
    noisy_image = np.copy(image)
    # Adding salt noise
    num_salt = np.ceil(salt_prob * image.size)
    coords = [np.random.randint(0, i - 1, int(num_salt)) for i in image.shape]
    noisy_image[coords[0], coords[1]] = 255
    
    # Adding pepper noise
    num_pepper = np.ceil(pepper_prob * image.size)
    coords = [np.random.randint(0, i - 1, int(num_pepper)) for i in image.shape]
    noisy_image[coords[0], coords[1]] = 0
    
    return noisy_image

# Define the iterative global thresholding function
def iterative_global_thresholding(image, initial_threshold=127, delta_T=1):
    T = initial_threshold
    diff = float('inf')
    
    while diff > delta_T:
        G1 = image[image > T]
        G2 = image[image <= T]
        
        m1 = np.mean(G1) if len(G1) > 0 else 0
        m2 = np.mean(G2) if len(G2) > 0 else 0
        
        new_T = 0.5 * (m1 + m2)
        diff = abs(T - new_T)
        T = new_T
    
    _, binary_image = cv2.threshold(image, T, 255, cv2.THRESH_BINARY)
    
    return binary_image, T

# Add noise to the original image
gaussian_noisy_image = add_gaussian_noise(image)
salt_pepper_noisy_image = add_salt_pepper_noise(image)

# Apply the iterative thresholding on the noisy images
binary_image_gaussian, final_threshold_gaussian = iterative_global_thresholding(gaussian_noisy_image)
binary_image_salt_pepper, final_threshold_salt_pepper = iterative_global_thresholding(salt_pepper_noisy_image)

# Display the results
print(f"Final Threshold for Gaussian Noise: {final_threshold_gaussian}")
print(f"Final Threshold for Salt-and-Pepper Noise: {final_threshold_salt_pepper}")

cv2.imshow('Original Image', image)
cv2.imshow('Gaussian Noisy Image', gaussian_noisy_image)
cv2.imshow('Salt and Pepper Noisy Image', salt_pepper_noisy_image)
cv2.imshow('Binary Image (Gaussian Noise)', binary_image_gaussian)
cv2.imshow('Binary Image (Salt-Pepper Noise)', binary_image_salt_pepper)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from skimage.util import random_noise

# Load the image in grayscale
image = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)

# Function to add noise using scikit-image's random_noise
def add_noise(image, noise_type='gaussian'):
    # Add noise of a specific type: 'gaussian', 'salt & pepper', etc.
    noisy_image = random_noise(image, mode=noise_type)
    # Convert the noisy image to the uint8 type and scale back to [0, 255]
    noisy_image = np.array(255 * noisy_image, dtype=np.uint8)
    return noisy_image

# Define the iterative global thresholding function
def iterative_global_thresholding(image, initial_threshold=127, delta_T=1):
    T = initial_threshold
    diff = float('inf')
    
    while diff > delta_T:
        G1 = image[image > T]
        G2 = image[image <= T]
        
        m1 = np.mean(G1) if len(G1) > 0 else 0
        m2 = np.mean(G2) if len(G2) > 0 else 0
        
        new_T = 0.5 * (m1 + m2)
        diff = abs(T - new_T)
        T = new_T
    
    _, binary_image = cv2.threshold(image, T, 255, cv2.THRESH_BINARY)
    
    return binary_image, T

# Add Gaussian and salt-and-pepper noise to the image
gaussian_noisy_image = add_noise(image, 'gaussian')
salt_pepper_noisy_image = add_noise(image, 's&p')

# Apply the iterative thresholding on the noisy images
binary_image_gaussian, final_threshold_gaussian = iterative_global_thresholding(gaussian_noisy_image)
binary_image_salt_pepper, final_threshold_salt_pepper = iterative_global_thresholding(salt_pepper_noisy_image)

# Display the results
print(f"Final Threshold for Gaussian Noise: {final_threshold_gaussian}")
print(f"Final Threshold for Salt-and-Pepper Noise: {final_threshold_salt_pepper}")

cv2.imshow('Original Image', image)
cv2.imshow('Gaussian Noisy Image', gaussian_noisy_image)
cv2.imshow('Salt and Pepper Noisy Image', salt_pepper_noisy_image)
cv2.imshow('Binary Image (Gaussian Noise)', binary_image_gaussian)
cv2.imshow('Binary Image (Salt-Pepper Noise)', binary_image_salt_pepper)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
